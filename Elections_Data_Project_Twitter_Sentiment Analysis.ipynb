```python
def clean_tweets(tweet):
    return ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)", " ", tweet).split()) 
```

```python
elections_df['text'] = elections_df["text"].apply(clean_tweets)
elections_df.head()
```

Resource: https://www.geeksforgeeks.org/twitter-sentiment-analysis-using-python/

```python
def tweet_sentiment(tweet): 
    # create TextBlob object of passed tweet text 
        analysis = TextBlob(clean_tweets(tweet)) 
        # set sentiment 
        if analysis.sentiment.polarity > 0: 
            return 'positive'
        elif analysis.sentiment.polarity == 0: 
            return 'neutral'
        else: 
            return 'negative'
```

```python
elections_df["tweet_sentiment"] = elections_df["text"].apply(tweet_sentiment)
elections_df.head()
```
