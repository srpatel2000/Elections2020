```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
end = pd.read_csv('endorsements-2020.csv')
end
```

```python
biden = end.loc[end['endorsee'] == "Joe Biden"]
sanders = end.loc[end['endorsee'] == "Bernie Sanders"]
warren = end.loc[end['endorsee'] == "Elizabeth Warren"]
bloomberg = end.loc[end['endorsee'] == "Michael Bloomberg"]
```

```python
biden_counts = biden.groupby(["points"]).count()
biden_list = biden_counts["date"].tolist()

sanders_counts = sanders.groupby(["points"]).count()
sanders_list = sanders_counts["date"].tolist()

warren_counts = warren.groupby(["points"]).count()
warren_list = warren_counts["date"].tolist()

bloomberg_counts = bloomberg.groupby(["points"]).count()
bloomberg_list = bloomberg_counts["date"].tolist()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# data to plot
n_groups = 7
bars1 = [11, 11, 5, 4]
bars2 = [11, 7, 11, 1]
bars3 = [45, 0, 13, 12]
bars4 = [6, 0, 3, 0]
bars5 = [5, 1, 1, 0]
bars6 = [3, 0, 0, 1]
bars7 = [0, 0, 0, 0]
```

```python
# set width of bar
barWidth = 0.3
 
# set height of bar
bars1 = biden_list
bars2 = sanders_list
bars3 = warren_list
bars4 = bloomberg_list
 
# Set position of bar on X axis
r1 = np.arange(len(bars1))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
r4 = [x + barWidth for x in r3]
r5 = [x + barWidth for x in r4]
r6 = [x + barWidth for x in r5]

# Make the plot
plt.bar(r1, bars1, color='#7f6d5f', width=barWidth, edgecolor='white', label='var1')
plt.bar(r2, bars2, color='#557f2d', width=barWidth, edgecolor='white', label='var2')
plt.bar(r3, bars3, color='#2d7f5e', width=barWidth, edgecolor='white', label='var3')
plt.bar(r4, bars4, color='#2d7f5e', width=barWidth, edgecolor='white', label='var4')

# Add xticks on the middle of the group bars
plt.xlabel('group', fontweight='bold')
plt.xticks([r + barWidth for r in range(4)], ['Biden', 'Sanders', 'Warren', 'Bloomberg'])
 
# Create legend & Show graphic
plt.legend()
plt.show()
```
